```python
import pandas as pd
import json

# Path file JSON dan output CSV
input_json = 'instagram_posts.json' 
output_csv = 'instagram_posts.csv'  

# Membaca data dari file JSON
with open(input_json, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Proses konversi data
flattened_data = []
for post in data:
    for comment in post.get("comments", []):
        flattened_data.append({
            "input_url": post.get("input_url"),
            "post_id": post.get("post_id"),
            "owner_username": post.get("owner_username"),
            "timestamp": post.get("timestamp"),
            "type": post.get("type"),
            "likes": post.get("likes"),
            "caption": post.get("caption"),
            "comment_id": comment.get("comment_id"),
            "text": comment.get("text"),
            "username": comment.get("username")
        })

# Membuat DataFrame dan menyimpan ke CSV
df = pd.DataFrame(flattened_data)
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"Data berhasil dikonversi dan disimpan di '{output_csv}'")
```
