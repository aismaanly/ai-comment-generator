```python
import pandas as pd

# Masukkan nama file CSV yang akan dikonversi
input_csv = 'DGSyIWjpLnt_Comments.csv'  # Ganti dengan nama file CSV Anda
output_json = 'DGSyIWjpLnt_Comments.json'  # Nama file JSON yang akan dibuat

try:
    # Membaca file CSV
    df = pd.read_csv(input_csv, sep=';')
    print(f"File '{input_csv}' berhasil dibaca.")

    # Menyimpan file sebagai JSON dengan memastikan emot tetap dipertahankan
    df.to_json(output_json, orient='records', indent=4, force_ascii=False)
    print(f"File JSON berhasil disimpan sebagai '{output_json}'.")

except FileNotFoundError:
    print(f"File '{input_csv}' tidak ditemukan. Pastikan file berada di lokasi yang benar.")
except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
